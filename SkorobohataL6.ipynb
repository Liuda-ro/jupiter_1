```python
import random
import json
import matplotlib.pyplot as plt
import numpy as np
```

1. Створити програму для генерації довільного масиву із 54 цілих чисел, які не перевищують по модулю 128, з наступними властивостями: 
1.1. При кожному запуску програми масив повинен бути різним.
1.2. Масив повинен містити 30 відсотків від’ємних елементів (розташованих в довільних місцях)

```python
rand = [random.choices([-random.randint(1,128), random.randint(0,128)], weights=[30,70])[0] for _ in range(54)]
print(rand)
```

2. Зберегти згенерований масив в файл. 

```python
try:
    with open("mas.txt", "w") as file:
        print("Файл створений;)")
        json.dump(rand, file)
        print("Записано масив")
except Exception as e:
    print(f"Помилка при створенні файлу: {e}")
```

3. Створити програму, яка зчитує із файлу згенерований масив попередньою програмою і виводить його на екран. 

```python
try:
    with open("mas.txt", "r") as file:
        print("Файл відкрито;)")
        rand = json.load(file)
        print(rand)
except Exception as e:
    print(f"Помилка при відкритті файлу: {e}")
```

4. Візуалізувати масив за допомогою стовпчастої горизонтальної діаграми. Виділити кольором мінімальний, другий мінімальний та третій мінімальний елементи, а також вказати ці значення на діаграмі. Виділити іншим кольором елементи із інтервалу [a,b], який задає користувач. 
5. Порахувати номер позиції третього мінімального елемента. 

```python
min_arr = []
print("Мінімальні елементи\n")
for _ in range(3): 
    min_elem_1 = min(rand)
    min_index_1 = rand.index(min_elem_1)
    print(min_elem_1)
    print(f"Індекс елемента {min_index_1}")
    min_arr.append(min_elem_1)
    rand.pop(min_index_1)


a = int(input("Введіть індекс початку інтервалу: "))
b = int(input("Введіть індекс кінця інтервалу: "))


interval_values = rand[a:b+1]
other_values = rand[:a] + rand[b+1:]


category_names = ['Min 1', 'Min 2', 'Min 3', 'Interval [a:b]', 'Other values']
results = {
    'Min 1': [min_arr[0]],
    'Min 2': [min_arr[1]],
    'Min 3': [min_arr[2]],
    'Interval [a:b]': interval_values,
    'Other values': other_values
}

def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()), dtype=object)
    category_colors = plt.get_cmap('RdYlGn')(np.linspace(0.15, 0.85, len(category_names)))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, max(len(values) for values in results.values()))

    for i, (label, color) in enumerate(zip(category_names, category_colors)):
        values = results[labels[i]]
        bar = ax.barh(label, len(values), color=color, height=0.5)
        ax.bar_label(bar, labels=[f"{len(values)} elements"], label_type='center')

    ax.legend(category_names, ncols=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(results, category_names)
plt.show()


```
